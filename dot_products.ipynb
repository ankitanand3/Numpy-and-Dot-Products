## Neuron: Vector x Vector

```python
import numpy as np
inputs = [1, 2, 3, 2.5]
weights = [0.2, 0.8, -0.5, 1.0]
bias = 2

output = np.dot(inputs, weights) + bias
print(output)

```

## Layer: Vector x Matrix

```python
inputs = [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]

layer_outputs = np.dot(weights, inputs) + biases
print(layer_outputs)
```

## Batch of Data: Matrix x Matrix

```python
inputs = [
    [1, 2, 3, 2.5],
    [2.0, 5.0, -1.0, 2.0],
    [-1.5, 2.7, 3.3, -0.8]
]
weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]

outputs = np.dot(inputs, np.array(weights).T) + biases
print(outputs)
```

## Two Layers

```python
inputs = [[1, 2, 3, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]

weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

weights2 = [[0.1, -0.14, 0.5],
            [-0.5, 0.12, -0.33],
            [-0.44, 0.73, -0.13]]
            
biases2 = [-1, 2, -0.5]

# convert lists to numpy arrays
inputs = np.array(inputs)
weights = np.array(weights)
biases = np.array(biases)
weights2 = np.array(weights2)
biases2 = np.array(biases2)

#  Calculate Layer 1 Outputs
layer1_outputs = np.dot(inputs, weights.T) + biases

# Calculate Layer 2 Outputs
layer2_outputs = np.dot(layer1_outputs, weights2.T) + biases2

print(layer2_outputs) 
```
